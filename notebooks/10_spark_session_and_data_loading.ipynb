# Spark Session and Data Loading

This notebook introduces the basic workflow for starting PySpark and loading a CSV file into a Spark DataFrame.

By the end of this notebook, you will understand how to:

- Import required libraries
- Check file encoding
- Preview data with Pandas
- Create a Spark session
- Load CSV data with PySpark
- Inspect Spark DataFrames

```python
# Import required libraries

import chardet
import pandas as pd

from pyspark.sql import SparkSession
```

### Package Definitions

- `chardet`: Detects file encoding before reading external data.
- `pandas`: Previews small datasets before loading them into Spark.
- `pyspark.sql.SparkSession`: Creates the main entry point for working with Spark DataFrames.

## 1. Check File Encoding

Before loading a CSV file, it can be useful to check its encoding. This helps avoid reading errors when working with external datasets.

```python
with open("../data/processed/test_file_01.csv", "rb") as file:
    raw_data = file.read()

encoding_result = chardet.detect(raw_data)
encoding_result
```

## 2. Preview Data with Pandas

Pandas is useful for quickly checking small datasets before loading them into Spark.

```python
pd.read_csv("../data/processed/test_file_01.csv")
```

## 3. Create a Spark Session

The Spark session is the main entry point for using PySpark. It allows us to create and work with Spark DataFrames.

```python
spark = (
    SparkSession.builder
    .appName("Session")
    .getOrCreate()
)

spark
```

## 4. Load CSV Without Header

By default, Spark does not automatically treat the first row as column names.

```python
df_pyspark = spark.read.csv("../data/processed/test_file_01.csv")

df_pyspark.show()
```

## 5. Load CSV With Header

Using `header=True` tells Spark to use the first row as column names.

```python
df_pyspark = spark.read.csv(
    "../data/processed/test_file_01.csv",
    header=True
)

df_pyspark.show()
```

## 6. Inspect the DataFrame

After loading the data, we can check the object type, preview records, and inspect the schema.

```python
type(df_pyspark)
```

```python
df_pyspark.head(3)
```

```python
df_pyspark.printSchema()
```

## Key Takeaways

- A Spark session is required before working with PySpark DataFrames.
- Spark can load CSV files with or without headers.
- `show()` displays rows in table format.
- `head()` returns records from the DataFrame.
- `printSchema()` shows column names and data types.
