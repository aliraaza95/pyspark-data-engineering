# PySpark DataFrame Basics

This notebook introduces the core operations used when working with PySpark DataFrames.

By the end of this notebook, you will understand how to:

- Load CSV data with schema inference
- Inspect DataFrame structure
- Select columns
- Check data types
- Generate summary statistics
- Add new columns
- Drop columns
- Rename columns

# Import Required Libraries

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
```

## Package Definitions

- `SparkSession`: Creates the entry point for working with Spark DataFrames.
- `pyspark.sql.functions`: Provides built-in functions for DataFrame transformations and column operations.

## 1. Create a Spark Session

A Spark session is required before working with PySpark DataFrames.

```python
spark = (
    SparkSession.builder
    .appName("DataFrame Basics")
    .getOrCreate()
)

spark
```

## 2. Load CSV With Schema Inference

Using `inferSchema=True` allows Spark to automatically detect column data types.

```python
df_pyspark = spark.read.csv(
    "../data/processed/test_file_01.csv",
    header=True,
    inferSchema=True
)

df_pyspark.show()
```

## 3. Inspect the DataFrame Structure

The schema shows column names and their detected data types.

```python
df_pyspark.printSchema()
```

## 4. Select a Single Column

Use `select()` to retrieve one column from the DataFrame.

```python
df_pyspark.select("Name").show()
```

## 5. Select Multiple Columns

Multiple columns can be selected by passing multiple column names.

```python
df_pyspark.select(
    "Name",
    "Age"
).show()
```

## 6. Check Data Types

The `dtypes` attribute returns column names and their corresponding data types.

```python
df_pyspark.dtypes
```

## 7. Generate Summary Statistics

The `describe()` method provides basic descriptive statistics for numerical columns.

```python
df_pyspark.describe().show()
```

## 8. Add a New Column

Use `withColumn()` to create a new column based on an existing one.

```python
df_with_bonus = df_pyspark.withColumn(
    "Age Plus Five",
    col("Age") + 5
)

df_with_bonus.show()
```

## 9. Drop a Column

The `drop()` method removes a column from the DataFrame.

```python
df_without_age = df_with_bonus.drop("Age Plus Five")

df_without_age.show()
```

## 10. Rename a Column

Use `withColumnRenamed()` to rename an existing column.

```python
df_renamed = df_pyspark.withColumnRenamed(
    "Age",
    "Employee Age"
)

df_renamed.show()
```

## Key Takeaways

- `inferSchema=True` automatically detects column data types.
- `printSchema()` displays the DataFrame structure.
- `select()` retrieves one or more columns.
- `dtypes` returns column data types.
- `describe()` generates summary statistics.
- `withColumn()` creates new columns.
- `drop()` removes columns.
- `withColumnRenamed()` renames existing columns.
